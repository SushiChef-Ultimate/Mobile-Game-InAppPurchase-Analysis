```python
# Pips and imports
!pip install numpy pandas matplotlib seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setup
MobileGame_DF = pd.read_csv("mobile_game_inapp_purchases.csv")
MobileGame_DF = MobileGame_DF.dropna(axis=0)
MobileGame_DF.describe()
```

Preprocessing:
    We set up MobileGame_DF as a pandas dataframe of "mobile_game_inapp_purchases.csv".
    our main preprocessing step is to drop null columns. Since nulls only make up around 2-5% of records, dropping them wont really affect our data, and help with ensuring that only properly collected and "complete" records make it into our visualizations.

```python
print(MobileGame_DF.columns)
```

```python
# Average spending by gender
plt.ylabel("Average In-App Purchases ($)")
sns.barplot(data=MobileGame_DF, x='Gender', y='InAppPurchaseAmount', hue='Gender')
```

```python
# 3 seperate bar graphs for each gender using catplot
x = sns.catplot(
    data=MobileGame_DF,
    x='SpendingSegment',
    kind='count',
    col='Gender', # Creates the 3 seperate barplots seperating by gender
    height=4,
    aspect=0.8,
    hue = 'Gender'
)

x.set_axis_labels("Spending Class", "# of Players")
```


IMPORTANT CONTEXT : People who spend money on mobile games are separated into 3 categories

    Minnow : Light and occasional spending
    Dolphin : Consistent and moderate spending
    Whale : Very frequent and heavy spending

```python
# Box plot for when first purchase is made after install by gender
plt.ylabel("Average Days Until First Purchase")
sns.boxplot(data=MobileGame_DF, x='Gender', y='FirstPurchaseDaysAfterInstall', hue='Gender')
```

```python
# Average spending by country
plt.figure(figsize=(18, 6))
plt.xticks(rotation=45)
sns.barplot(data=MobileGame_DF, x='Country', y='InAppPurchaseAmount', hue='Country')
plt.ylabel("Average In-App Purchases ($)")
```

```python
# Average spending by genre
plt.figure(figsize=(16, 6))
plt.xticks(rotation=30)
sns.barplot(data=MobileGame_DF, x='GameGenre', y='InAppPurchaseAmount', hue='GameGenre')
plt.xlabel("Game Genre")
plt.ylabel("Average In-App Purchases ($)")
```

```python
# Box plot for when first purchase is made after install by game genre
plt.figure(figsize=(16, 6))
plt.xticks(rotation=30)
sns.boxplot(data=MobileGame_DF, x='GameGenre', y='FirstPurchaseDaysAfterInstall', hue='GameGenre')
plt.xlabel("Game Genre")
plt.ylabel("Average Days Until First Purchase")
```
